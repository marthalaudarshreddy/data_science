```python
#cummulative distibutive funcitons(CDF)(gives us that from before upto that point how much percentage is present)
```

```python
#differencitae cdf you will get pdf and if you integrate pdf you will get cdf
```

```python
plt.close()
counts,edges=np.histogram(setosa["petal_length"],bins=10,density=True)#bins will give no of bars of eqaual intervel
pdf = counts/ (sum(counts))
cdf=np.cumsum(pdf)
plt.plot(edges[1:],pdf)
plt.plot(edges[1:],cdf)
plt.show()
```

```python
counts,edges=np.histogram(setosa["petal_length"],bins=10,density=True)#bins will give no of bars of eqaual intervel
pdf = counts/ (sum(counts))
cdf=np.cumsum(pdf)
plt.plot(edges[1:],pdf)
plt.plot(edges[1:],cdf)

counts,edges=np.histogram(virginica["petal_length"],bins=10,density=True)#bins will give no of bars of eqaual intervel
pdf = counts/ (sum(counts))
cdf=np.cumsum(pdf)
plt.plot(edges[1:],pdf)
plt.plot(edges[1:],cdf)

counts,edges=np.histogram(versicolor["petal_length"],bins=10,density=True)#bins will give no of bars of eqaual intervel
pdf = counts/ (sum(counts))
cdf=np.cumsum(pdf)
plt.plot(edges[1:],pdf)
plt.plot(edges[1:],cdf)

plt.show()


#the usefull information here is of CDF is at 5 ie less than this 95 percent i am correct that it is brown and 10
# percent wrong that it is not red and we can extract usefull information
```
